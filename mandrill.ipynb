```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
from skimage.feature import corner_peaks,corner_harris
from skimage.exposure import equalize_hist
from skimage.color import rgb2gray

mandrill=io.imread('C:/Users/thefl/Downloads/mandrill.png')
print(mandrill.shape)
mandrill=equalize_hist(rgb2gray(mandrill))
print(mandrill.shape)
# mandrill1=mandrill
# plt.imshow(mandrill1)
```

```python
# mandrill1=equalize_hist(mandrill)
# plt.imshow(mandrill1)
corners= corner_peaks(corner_harris(mandrill),min_distance=2)
# print(corners)
def show_corners(corners,images):
    fig=plt.figure()
    plt.gray()
    plt.imshow(images)
    y_corner,x_corner=zip(*corners)
    plt.plot(x_corner,y_corner,'or')
    plt.xlim(0,images.shape[1])
    plt.ylim(images.shape[0],0)
    fig.set_size_inches(np.array(fig.get_size_inches())*1.5)
show_corners(corners,mandrill)
```
